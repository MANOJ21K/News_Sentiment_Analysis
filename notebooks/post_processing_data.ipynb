```python
from main import NewsDataUnstructured
```

```python
# 2. Create an instance of the class with the desired ticker
ticker = 'AAPL'
news_instance = NewsDataUnstructured(ticker)

# 3. Process the data
news_instance.process_data()

# 4. Retrieve the processed data
processed_data = news_instance.get_processed_data()

# Print the processed DataFrame
processed_data
```

```python
from datetime import datetime

df = processed_data[['published_date', 'Description']]
# After importing pandas and datetime
df['Date'] = df['published_date'].apply(lambda x: datetime.strptime(x, "%a, %d %b %Y %H:%M:%S %Z"))

# To avoid the SettingWithCopyWarning, use .loc to modify the DataFrame explicitly
df.loc[:, 'Date'] = df['Date']

# Print the DataFrame with the 'Date' column
df = df.iloc[:, 1:]
```

```python
from summarizer import Summarizer

df['Final_Description'] = ""
model = Summarizer()
for i in range(len(df)):
    result = model(df['Description'][i], min_length=10, max_length=512)
    df['Final_Description'][i] = result
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline
model = BertForSequenceClassification.from_pretrained("ahmedrachid/FinancialBERT-Sentiment-Analysis",num_labels=3)
tokenizer = BertTokenizer.from_pretrained("ahmedrachid/FinancialBERT-Sentiment-Analysis")
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)


df['Sentiment'] = ''
for i in range(len(df)):
    results = nlp(df['Final_Description'][i])
    df['Sentiment'][i] = results
    
df['Label'] = df['Sentiment'].apply(lambda x: x[0]['label'])
df['Score'] = df['Sentiment'].apply(lambda x: x[0]['score'])
df = df.drop('Sentiment', axis=1)
```

```python
import seaborn as sns
sns.countplot(data=df, x='Label')
```

```python
df.to_csv("Final.csv")
```
